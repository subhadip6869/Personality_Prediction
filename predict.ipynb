```python
from keras.models import load_model
model = load_model("personality.h5")
```

```python
classes = {
    'Agreeableness': 0,
    'Conscientiousness': 1,
    'Extraversion': 2,
    'Neuroticism': 3,
    'Openness': 4
}
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
org_img = cv2.imread("./dataset/dataset/test_set/Conscientiousness/IMG_20200215_174850.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(org_img, cmap="gray")
plt.show()
org_img = cv2.resize(org_img, (600, 600))
test_input = org_img.reshape((1, 600, 600, 1))
```

```python
predict = model.predict(test_input)
predict_key = np.argmax(predict, axis=1)[0]

class_keys = list(classes.keys())
class_vals = list(classes.values())

class_keys[predict_key]
```
